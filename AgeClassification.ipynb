Função em python que lê como input uma idade e retorna a classificaçao da idade do indivíduo da seguinte forma:

Criança – 0 a 12 anos

Adolescente – 13 a 17 anos

Adulto – acima de 18 anos

Se a idade for um número negativo, mostra a mensagem "Idade inválida".

```python
def classifica_idade(idade):
  if idade >= 18:
    print('A classificação dessa idade é Adulto')
  elif idade >= 13:
    print('A classificação dessa idade é Adolescente')
  elif idade >= 0:
    print('A classificação dessa idade é Criança')
  else:
    print('Idade inválida')
```

```python
idade = float(input('Digite sua idade para saber a classificação dela: '))

classifica_idade(idade)
```
