Função em python que recebe como input um determinado horário em horas e retorna a conversão deste horário para segundos.

```python
def time_converter(t):
  hh, mm, ss = t.split(':')
  return int(hh) * 3600 + int(mm) * 60 + int(ss)
```

```python
t = input('Digite um horário no formato HH:MM:SS para saber sua equivalência em segundos: ')

print('Essa hora equivale a {0} segundos.'.format(time_converter(t)))
```
