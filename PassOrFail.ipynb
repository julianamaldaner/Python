Função em python que calcula a média artimética de uma aluna que cursou a disciplina de Programação I. A função recebe como input as notas M1, M2 e M3 e calcula a média. Após a média calculada, a função anuncia se a aluna foi aprovada, reprovada ou pegou exame, de acordo com a seguinte classificação:

* Se a média estiver entre 0.0 e 4.0, a aluna está reprovada;
* Se a média estiver entre 4.1 e 6.0, a aluna pegou exame;
* Se a média for maior do que 6.0, a aluna está aprovada;
* Se a aluna pegou exame, é lida a nota do exame. Se a nota do exame for maior do que 6.0, está aprovada, se não, está reprovada.

```python
def calcula_media(M1,M2,M3):
  media = (M1 + M2 + M3) / 3
  if media > 6.0:
    print('Parabéns! Você está aprovado(a).')
  elif media >= 4.1:
    exame = float(input('Sua média é {0:.1f}, por isso você pegou exame. Digite a nota do seu exame: '.format(media)))
    if exame > 6.0:
      print('Parabéns! Você está aprovado(a).')
    else:
      print('Infelizmente você está reprovado(a).')
  else:
    print('Infelizmente você está reprovado(a).')
```

```python
M1 = float(input('Digite a primeira nota recebida (M1): '))
M2 = float(input('Digite a segunda nota recebida (M2): '))
M3 = float(input('Digite a terceira nota recebida (M3): '))

calcula_media(M1, M2, M3)
```
