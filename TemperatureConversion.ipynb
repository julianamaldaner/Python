Função em Python que lê uma temperatura em graus Celsius e retorna a conversão dela em graus Fahrenheit.

```python
def conversor_temperatura(temperatura):
  F = (9 * temperatura + 160) / 5
  print('{0:.1f} graus Celsius equivalem e {1:.1f} graus Fahrenheit.'.format(temperatura, F))
```

```python
temperatura = float(input('Informe uma temperatura em graus Celsius para saber sua equivalência em graus Fahrenheit: '))

conversor_temperatura(temperatura)
```
