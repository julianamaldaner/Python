Função que atribui cada uma das alunas em uma das equipes da lista "equipes", de modo que cada equipe pode ter 3 alunas no máximo. 

A distribuição é feita por ordem alfabética e as equipes são colocadas em um dicionário Python.

```python
def ordenacao(alunas, equipes):
  ### escrever abaixo
  alunas.sort()
  key_list = equipes
  i = 0
  a = 0

  equipes_dict = {key: [] for key in key_list}

  for aluna in alunas:
    if i == 3:
      a += 1
      i = 0
    
    equipes_dict[list(equipes_dict.keys())[a]].append(aluna)
    i += 1

  print(equipes_dict)
```

```python
alunas = ['Maria', 'Ana','Camila','Mariana','Elaine','Patricia','Marina','Erica','Larissa','Luiza', 'Nicole','Bruna']
equipes = ['Equipe 1', 'Equipe 2', 'Equipe 3', 'Equipe 4']

ordenacao(alunas, equipes)
```
