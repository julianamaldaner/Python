Função em Python que conta quantas vezes uma letra aparece em cada elemento da lista. O resultado é mostrado em outra lista.

```python
def contador(user_letter, alunas):
  letter_count = 0
  letters_list = []

  for elemento in alunas:
    elemento = elemento.lower()
    letter_count = elemento.count(user_letter.lower())
    letters_list.append(letter_count)

  print('O número de ocorrências da letra {0} em cada elemento da lista é {1}'.format(user_letter,letters_list))
```

```python
user_letter = input('Digite uma letra: ')
alunas = ['Maria', 'Ana','Camila','Mariana','Elaine','Patricia','Marina','Erica','Larissa','Luiza', 'Nicole','Bruna']

contador(user_letter, alunas)
```
