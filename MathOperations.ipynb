```python
def operations(a,b):
  adicao = a + b
  print('O resultado da adicao dos dois números inteiros é {0}'.format(adicao))
  subtracao = a - b
  print('O resultado da subtracao dos dois números inteiros é {0}'.format(subtracao))
  multiplicacao = a * b
  print('O resultado da multiplicacao dos dois números inteiros é {0}'.format(multiplicacao))
  
  if b == 0:
    print('Impossível chegar a um resultado na divisão, no segundo número, informe um que seja diferente de 0')
  else:
    divisao = a / b
    print('O resultado da divisao dos dois números inteiros é {0:.1f}'.format(divisao))
```

```python
a = int(input('Digite um número inteiro: '))
b = int(input('Digite outro número inteiro: '))

operations(a,b)
```
